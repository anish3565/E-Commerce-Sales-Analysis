```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset=pd.read_excel("Superstore_USA.xlsx")
```

```python
dataset['Product Base Margin'].fillna(dataset['Product Base Margin'].mean(), inplace=True)
```

```python
dataset.isnull().sum()
```

## Order Priority

```python
dataset['Order Priority'].value_counts()
```

```python
dataset['Order Priority'].unique()
```

Performing Data cleaning, as we have 'Critical' and 'Critical ' both

```python
dataset['Order Priority'].replace("Critical ","Critical")
```

```python
dataset['Order Priority']=dataset['Order Priority'].replace("Critical ","Critical")
```

```python
dataset['Order Priority'].value_counts()
```

## Graphical Analysis

```python
# Simple graphing code
sns.countplot(x="Order Priority", data=dataset)
plt.show()
```

```python
plt.figure(figsize=(4,3))
sns.countplot(x="Order Priority", data=dataset)
plt.title("Count of Order Priority")
plt.savefig("Count of Order Priority.jpg")
plt.show()
```

## Ship Mode

```python
dataset['Ship Mode'].value_counts()
```

```python
dataset['Ship Mode'].value_counts().index
# Can act as X-axis
```

```python
dataset['Ship Mode'].value_counts().values
# Can act as Y-axis
```

```python
x=dataset['Ship Mode'].value_counts().index
y=dataset['Ship Mode'].value_counts().values
```

```python
plt.figure(figsize=(5,4))
plt.pie(y, labels=x,startangle=60,autopct="%0.2f%%")
plt.legend(loc=3)
plt.show()
```

```python
plt.figure(figsize=(5,4))
sns.countplot(x="Ship Mode", data=dataset)
plt.show()
```

For comparative analysis as pairplot of say 2 features like "Ship Mode" and "Product Category", we use **hue**

```python
plt.figure(figsize=(6,4))
sns.countplot(x="Ship Mode", data=dataset, hue="Product Category")
plt.show()
```
